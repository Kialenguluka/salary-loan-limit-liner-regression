```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Importando o Ficheiro
Base_Dados = pd.read_excel('BaseDados_RegressaoLinear.xlsx', 'Plan1')
```

```python
#Analise Exploratoria
Base_Dados.info()
print('='*50)
Base_Dados.describe()
print('='*50)
Base_Dados.head()
```

```python
Base_Dados.dtypes
Base_Dados.columns
```

```python
#Convertendo a tabela para array
Eixo_x = Base_Dados.iloc[:,0].values
Eixo_y = Base_Dados.iloc[:,1].values
```

```python
Eixo_x
```

```python
Eixo_y
```

```python
#Analise Grafica
plt.figure(figsize=(10,5))
plt.scatter(Eixo_x, Eixo_y)
plt.title('Grafico com 2 eixos [Salario x Limite de Emprestimo]' )
plt.xlabel('Salario')
plt.ylabel('Limite de Emprestimo');
```

```python
#Verificando se ha valores nulos graficamente
plt.figure(figsize=(10, 5))
sns.heatmap(Base_Dados.isnull());
```

```python
#Graficos de 2 eixos comparando cada coluna com ela mesma e com outras.
#AVISO: nao usar para base de dados com muitas colunas.
sns.pairplot( Base_Dados );
```

```python
#Calculando a Correlacao entre Salario x Limite de Emprestimos usando a funcao corrcoef()
Correl = np.corrcoef(Eixo_x, Eixo_y)
print('Correlacao: ')
Correl
```

```python
#Visualizando grafica da Correlacao
plt.figure(figsize=(10, 5))
sns.heatmap(Correl, annot=True) # O annot mostra os valores no meio do grafico
```

```python
#Convertendo o array para a matriz. O comando reshape converte a lista em matriz e o (-1,1) fara com que mexa apenas nas colunas.
Eixo_x = Eixo_x.reshape(-1, 1)
Eixo_y = Eixo_y.reshape(-1, 1)
```

```python
#Trabalhando no modelo - Treinamento
#Importando a biblioteca
from sklearn.model_selection import train_test_split
```

```python
#Dividir os dados entre Dados de Teste e de Dados de Treinamento
#Divisao Aleatoria
x_treinamento, x_teste, y_treinamento, y_teste = train_test_split(
    Eixo_x,
    Eixo_y,
    test_size=0.20
)
```

```python
print( len(x_treinamento), len(x_teste) ) 
```

```python
#Treinamento do Modelo
#Importando a biblioteca
from sklearn.linear_model import LinearRegression
```

```python
#Definindo uma variavel para armazenar a formula da Regressao Linear
FuncaoRegressao = LinearRegression()
```

```python
#Usando o a funcao fit() para aplicar os calculos estatisticos.
# Treinando o modelo
FuncaoRegressao.fit(x_treinamento, y_treinamento)
```

```python
#Usando a funcao score() para testar a perfomance do modelo. A funcao calcula o quanto as variaveis se relacionam
FuncaoRegressao.score(x_treinamento, y_treinamento)
```

```python
#Tracando a Recta da Regressao usando a funcao predict() responsavel pelas previsoes.
plt.figure( figsize=(10,5) )
plt.scatter(x_treinamento, y_treinamento)
plt.plot(x_teste, FuncaoRegressao.predict(x_teste), color='red');
```

```python
#Calculando o Erro nas previsoes
Previsoes = FuncaoRegressao.predict( x_teste )

from sklearn import metrics

print('RMSE: ', np.sqrt(metrics.mean_squared_error(y_teste, Previsoes) ) )
```
